## Building the model

```python
income.to_excel(r'C:\Users\femif\Downloads\dataincome.xlsx')
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
```

```python
categorical_cols = income.select_dtypes(include=['object']).columns
income_encoded = pd.get_dummies(income, columns=categorical_cols)
```

```python
income_encoded
```

```python
X = income_encoded.drop('Income_ <=50K', axis=1)
y = income_encoded['Income_ <=50K'] 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
model = LogisticRegression()
```

```python
model
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred 
```

```python
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
print(classification_report(y_test, y_pred))
```
