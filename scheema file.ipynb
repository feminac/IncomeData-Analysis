# Dataset name: Income Prediction Dataset
Description: A dataset containing features related to individuals and their income levels.

features:
  - name: age
  
    type: numerical
    
    description: Age of the individual.
    
  - name: education_level
  
    type: categorical
    
    description: Highest level of education achieved.
    
  - name: occupation
  
    type: categorical
    
    description: Type of occupation.
    
  - name: marital_status
  
    type: categorical
    
    description: Marital status of the individual.
    
  - name: relationship
  
    type: categorical
    
    description: Relationship status of the individual.
    
  - name: race
  
    type: categorical
    
    description: Ethnicity of the individual.
    
  - name: gender
  
    type: categorical
    
    description: Gender of the individual.
    
  - name: hours_per_week
  
    type: numerical
    
    description: Number of hours worked per week.
    
  - name: native_country
  
    type: categorical
    
    description: Country of origin.
    
  - name: income
  
    type: categorical
    
    description: Target variable indicating income level (<=50K or >50K).
